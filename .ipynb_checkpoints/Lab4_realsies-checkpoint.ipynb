```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
#import seaborn

data_xvals = np.atleast_2d( [ 1.0, 3.0, 5.0, 6.0, 7.0, 8.0 ] )
data_yvals = np.sin( data_xvals )


def Klin(x, y, noise):
    n = 0
    if noise:
        n = 1 if x == y else 0
    return x*y + n*.1


def Kpoly(x,y, noise):
    n = 0
    if noise:
         n = 1 if x == y.tolist() else 0
    return (1+(x*y))**3 + n*.1

def Kiso(x,y, noise):
    n = 0
    if noise:
         n = 1 if x == y.tolist() else 0
    return np.exp(-.5*(x-y)**2) + n*.1

def creatematrix(rows, columns, Kernel, noise):
    k = np.zeros((len(rows), len(columns)))
    for i in xrange(len(rows)):
        for j in xrange(len(columns)):
            k[i,j] = Kernel( rows[i], columns[j], noise )
    return k

def Plot(x_predict, mean, sigma, data_xvals, data_yvals, title):
    
    # Reshape all the mean, sigma, and predict
    a = np.reshape(x_predict, (1,121))
    b = np.reshape(mean,(1,121))
    c = np.reshape(sigma, (1, 121))

    plt.figure()
    plt.scatter(data_xvals[0],data_yvals[0], s=10)
    plt.plot(a[0], b[0],label='mean')
    
    # Reshape all the matricies 
    a = np.reshape(x_predict, (121,1))
    b = np.reshape(mean,(121,1))
    c = np.reshape(sigma, (121, 1))
    
    # Fill between the +- sigma 
    plt.fill(np.concatenate([a, a[::-1]]),
         np.concatenate([b - c,
                        (b + c)[::-1]]),
         alpha=.3, fc='b', ec='None', label='variance')
    
    plt.grid(True, color = "gray")
    plt.title(title)
    plt.legend(loc='upper left',frameon=False)
    plt.show()





mkplot( Klin, 0, "Linear Noiseless")
mkplot( Klin, 1, "Linear Noisy")
mkplot( Kpoly, 0, "Polynomial Noiseless")
mkplot( Kpoly, 1, "Polynomial Noisy")
mkplot( Kiso, 0, "Isometric Gaussian Noiseless")
mkplot( Kiso, 1, "Isometric Gaussian Noisy")



```

```python
def mkplot(kernel, noise, title):
    t = np.arange(-2, 10.01, .1)
    
    #Collect each of the matricies (K(x,x), K(x, xtest), K(xtest,x), 
    # and K(xtest, xtest))
    kxx = creatematrix( data_xvals[0], data_xvals[0], kernel, noise)
    ktx = creatematrix( t, data_xvals[0], kernel, noise )
    kxt = creatematrix( data_xvals[0], t, kernel, noise )
    ktt = creatematrix( t, t, kernel, noise )
    kxxinv = np.linalg.pinv(kxx)
    
    # Find each of the individual Mu and Sigmas for the matricies
    muk = np.dot(np.dot(ktx,kxxinv),np.transpose(data_yvals))
    sik = ktt - np.dot(np.dot(ktx, kxxinv), kxt)

    # Plot each of the graphs 
    Plot(t, muk, sik.diagonal(), data_xvals, data_yvals, title)
```
